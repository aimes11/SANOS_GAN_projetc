# AAPL 

```python
aapl = pd.read_csv('aapl_2021_2023.csv')
aapl.head()
```

```python
print("Before:", aapl.columns.tolist())

aapl.columns = aapl.columns.str.strip().str.replace(r"[\[\]]", "", regex=True)

print("After:", aapl.columns.tolist())

aapl["QUOTE_DATE"] = pd.to_datetime(aapl["QUOTE_DATE"], errors="coerce")
print(aapl["QUOTE_DATE"].dropna().sort_values().drop_duplicates().tail(20))
```
